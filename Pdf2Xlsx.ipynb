```python
import os
import numpy as np
import pandas as pd

from tqdm import tqdm
import glob
import regex as re
import multiprocessing
import tabula
import xlrd
import pdfplumber
import multiprocessing
import warnings
warnings.simplefilter("ignore")
import tabula

pd.set_option("display.max_rows", None) 
pd.set_option("display.max_columns", None)
```

```python
files=glob.glob('T-242 REPORT/Batch Report/All/*.pdf')
files=[file for file in files if '~' not in file]
print("# of files to be processed ", len(files))
```

```python
df  = pd.DataFrame()
count = 0
for file in tqdm(files):
    tabula.convert_into(file, "sample.csv", output_format="csv", pages='all')
    
    data = pd.read_csv("sample.csv")
    labels = str(data.iloc[[28],[0]])

    before = labels.index('Product Code:')
    after = labels.index('Batch Size:')
    bfg = labels[before + len('Product Code:') : after]

    before = labels.index('Batch Number:')
    after = labels.index('Lot Number')
    batch = labels[before + len('Batch Number:') : after]

    before = labels.index('Lot Number:')
    after = labels.index('Operator Name')
    lot = labels[before + len('Lot Number:') : after]
    lot = lot[0]

    data = pd.read_csv("sample.csv",header = 1)
    data = data.dropna()
    data.drop(data[ data['TIME'] == 'TIME' ].index, inplace = True)

    data['bfg_code'] = bfg
    data['batch_id'] = batch
    data['lot'] = lot

    df = pd.concat([df, data], axis=0)
    count = count+1

df.to_excel('T242.xlsx', index=False)
print('done')
```
